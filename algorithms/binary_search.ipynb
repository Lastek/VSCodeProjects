## Binary Search
With binary search we begin the search in the middle of the array \
The target value must be somewhere between the beginning and end of the array \
We're also assuming that the <u> __array is sorted first.__</u>

We construct two pointers *L* and *R*:
- *L* begins on the left and moves right 
- *R* begins on the right and moves left 

Now lets say we have an `array` of `array.size == 51`

Now if we pick the middle of the array then our target will \
either be `> target` or `< target` or ` == target`




Since the center value is less than 20 then we segment again \
Our new middle is between the center and *R*
We bounce between each segment till we find our value
 

If the pointers move past eachother

then it means our target does not exist in the array!

$$\text{Time Complexity}$$
$$ 40 \rightarrow 20 \rightarrow 10 \rightarrow ... \rightarrow 1 $$
$$n \rightarrow \frac{n}{2}\rightarrow \frac{n}{4} \rightarrow \frac{n}{8}\rightarrow ... \approx 1$$
$$n \rightarrow \frac{n}{2}\rightarrow \frac{n}{2^2}\frac{n}{2^3}\rightarrow ... \rightarrow \frac{n}{2^x} \approx 1$$
$$ 2^x \approx n $$
$$ x \approx log_2{n} $$
$$ O(log_2{n}) = O(log{n}) $$

````python
        
def binary_search(arr, target): #time complexity of O(log(n))
    left = 0
    right = len(arr) -1
    while left <= right:
        mid = (left + right) // 2 #round it down too if you need to (floor)
        if arr[mid] == target:  #incase we land on value in middle
            return mid 
        elif target < arr[mid]:
            right = mid - 1
        else:
            left = mid + 1
        return -1   #if it doesn't exist
    
````
