```python
def decoder(url:str) -> None:
    # Read the html table into a dataframe
    df = pd.read_html(url, encoding='utf-8', header=0) 
    # Get the first table
    df = df[0] 

    # Calculate grid size based on maximum x and y coordinates in the DataFrame
    max_x = df['x-coordinate'].max() + 1  
    max_y = df['y-coordinate'].max() + 1

    # Create a blank grid with the calculated size
    grid = [[' ' for _ in range(max_x)] for _ in range(max_y)]

    # Loop over the rows of the DataFrame
    for index, row in df.iterrows():
        # Get the x and y coordinates and the character
        x = row['x-coordinate']
        y = row['y-coordinate']
        character = row['Character']
        
        # Place the character in the right position on the grid
        grid[y][x] = character

    # Print the grid in reverse order
    for row in reversed(grid):
        print(''.join(row)) # join the row into a string and print it

decoder(url)
```
