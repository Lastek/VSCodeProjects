**Insertion Sort** is similar to **Selection Sort** in the aspect of how it treats the data. If we have an array the algorightm has two parts: *unsorted* and *sorted*. We also use a temp variable to swap out values. However, insertion sort is sligly better when the data is partially sorted. Fewer swaps but more comparisons

### Runtime complexity $O(n^2)$ for unsorted list
### Runtime complexity $O(n)$ for partially sorted list

We use `i` and `j` variables with two loops to iterate through an array or list.
`i` starts at 1 and moves up within an outer loop. `j` is equal to `i` and `j` is controlled by an inner loop. The inner loop

As `j` scans thought the array backwards, it swaps a value of the current element with the one behind it if the current value is less than the one behind it. If the case `n[j] < n[j-1]` holds then `n[j]` is placed into temp and `n[j-1]` is moved into the current spot. the `temp` element is moved into `n[j-1]`.


### Bench result:
1000 random elements 1-100 \
3.81 s ± 18.3 ms per loop (mean ± std. dev. of 10 runs, 10 loops each)

```python
def fill_list(n=10000) -> list[int]:
    import random
    random_ints = [random.randint(1, 100) for _ in range(n)]
    return random_ints

unsorted_list: list[int] = fill_list()

sorted_list = unsorted_list.copy()

def insertion_sort(a) -> None:
    for i in range(1, len(a)):
        j=i
        while j > 0 and a[j] < a[j-1]:
            k=j-1
            temp = a[j]
            a[j] = a[k]
            a[k] = temp
            j -= 1

insertion_sort(sorted_list)
#doen sorting
print("done sorting")

#basic length test
def test_lists_lengths(a: list[int], b:list[int])-> bool:
    if len(a) == len(b): return True
    else: return False

#basic function to make sure that the values in one list occur the exact number of times
def test_lists_count(a: list[int], b:list[int])-> bool: # type: ignore
    for i in range(0, len(sorted_list)):
        if a.count(a[i]) == b.count(a[i]):
            return True
        else:
            return False

# relies on test_lists_count passing successfully
def test_lists_contents(a: list[int], b:list[int])-> bool:
    for i in range(0, len(a)):
        #get count of items from list a for current index
        #make sure the item occurs n times in list b
        a_c = a.count(a[i])
        b_c = 0
        
        #run through list b finding all matching values
        # increment the b_c counter on match
        for j in range(0, len(a)):
            if a[i] == b[j] : b_c += 1
        #if there is any mismatch then immediately fail it
        if a_c != b_c: return False
 
    return True

#Suite main func
def test_suite_run(a: list[int], b:list[int])-> bool:
    success = False
    
    try: 
        success = test_lists_lengths(a,b)
        if success == False: raise Exception()
    except: print("Exception: test_lists_lengths")
    else: print("test_lists_lengths : Pass = ", success )
    
    try:
        success = test_lists_count(a,b)
        if success == False: raise Exception()
    except: print("Exception: test_lists_count")
    else: print("test_lists_count: Pass = " , success)
    
    try:        
        success = test_lists_contents(a,b)
        if success == False: raise Exception()
    except: print("Exception: test_lists_contents")
    else: print("test_lists_contents: Pass = " , success)
    
    return success

test_passed = test_suite_run(sorted_list, unsorted_list)
print("test_suite: Pass = " , test_passed)

if test_passed: 
    %timeit -r 4 -n 3 insertion_sort(fill_list())

#run test
```
